```python
import cv2
import matplotlib.pyplot as plt

# Load Haar Cascade classifiers
face_classifier = cv2.CascadeClassifier('/Users/myyntiimac/Desktop/Hercasscade classifier/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('/Users/myyntiimac/Desktop/Hercasscade classifier/haarcascade_eye.xml')

# Load the image
img = cv2.imread('/Users/myyntiimac/Desktop/image read/web.png')


# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_classifier.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Loop through detected faces and eyes
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    
    eyes = eye_classifier.detectMultiScale(roi_gray)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)

# Convert BGR image to RGB for displaying with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display result using matplotlib
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
